```python
from flask import Flask, request, jsonify
import pickle
import numpy as np
from pyngrok import ngrok
```

```python
model = pickle.load(open("fraud_model.pkl", "rb"))
```

```python
app = Flask(__name__)
```

```python
@app.route('/')
def home():
    return "UPI Fraud Detection Backend Running"
```

```python
@app.route('/predict', methods=['POST'])
def predict():

    data = request.get_json()

    features = np.array([[
        data['Merchant_ID'],
        data['Transaction_Type'],
        data['Payment_Gateway'],
        data['Transaction_City'],
        data['Transaction_State'],
        data['Transaction_Status'],
        data['Device_OS'],
        data['Transaction_Frequency'],
        data['Merchant_Category'],
        data['Transaction_Channel'],
        data['Transaction_Amount_Deviation'],
        data['Days_Since_Last_Transaction'],
        data['amount']
    ]])

    prediction = model.predict(features)[0]

    if prediction == 1:
        result = "Fraud Transaction"
    else:
        result = "Legitimate Transaction"

    return jsonify({
        "prediction": result
    })
```
