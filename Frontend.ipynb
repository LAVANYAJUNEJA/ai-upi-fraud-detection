```python
import gradio as gr
import pickle
import numpy as np
```

```python
model = pickle.load(open("fraud_model.pkl", "rb"))

try:
    scaler = pickle.load(open("scaler.pkl", "rb"))
except:
    scaler = None
```

```python
def predict(
     Merchant_ID,
Transaction_Type,
Payment_Gateway,
Transaction_City,
Transaction_State,
Transaction_Status,
Device_OS,
Transaction_Frequency,
Merchant_Category,
Transaction_Channel,
Transaction_Amount_Deviation,
Days_Since_Last_Transaction,
amount
):

    data = np.array([[
        Merchant_ID,
Transaction_Type,
Payment_Gateway,
Transaction_City,
Transaction_State,
Transaction_Status,
Device_OS,
Transaction_Frequency,
Merchant_Category,
Transaction_Channel,
Transaction_Amount_Deviation,
Days_Since_Last_Transaction,
amount
    ]])

    prediction = model.predict(data)[0]

    if prediction == 1:
        return " FRAUD DETECTED"
    else:
        return "SAFE TRANSACTION"
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as demo:

    gr.Markdown("UPI Fraud Detection System")
    gr.Markdown(" Enter transaction details to check fraud")

    with gr.Row():

        with gr.Column():
            Merchant_ID = gr.Number(label="Merchant ID")
            Transaction_Type = gr.Number(label="Transaction Type")
            Payment_Gateway = gr.Number(label="Payment Gateway")
            Transaction_City = gr.Number(label="Transaction City")
            Transaction_State = gr.Number(label="Transaction State")
            Transaction_Status = gr.Number(label="Transaction Status")
            Device_OS = gr.Number(label="Device OS")

        with gr.Column():
            Transaction_Frequency = gr.Number(label="Transaction Frequency")
            Merchant_Category = gr.Number(label="Merchant Category")
            Transaction_Channel = gr.Number(label="Transaction Channel")
            Transaction_Amount_Deviation = gr.Number(label="Transaction Amount Deviation")
            Days_Since_Last_Transaction = gr.Number(label="Days Since Last Transaction")
            amount = gr.Number(label="Transaction Amount")

    output = gr.Textbox(label=" Prediction Result")

    with gr.Row():
        predict_btn = gr.Button(" Predict", variant="primary")
        clear_btn = gr.Button("Clear")

    predict_btn.click(
        fn=predict,
        inputs=[
            Merchant_ID,
            Transaction_Type,
            Payment_Gateway,
            Transaction_City,
            Transaction_State,
            Transaction_Status,
            Device_OS,
            Transaction_Frequency,
            Merchant_Category,
            Transaction_Channel,
            Transaction_Amount_Deviation,
            Days_Since_Last_Transaction,
            amount
        ],
        outputs=output
    )

    clear_btn.click(
        fn=lambda: ("",)*14,
        inputs=[],
        outputs=[
            Merchant_ID,
            Transaction_Type,
            Payment_Gateway,
            Transaction_City,
            Transaction_State,
            Transaction_Status,
            Device_OS,
            Transaction_Frequency,
            Merchant_Category,
            Transaction_Channel,
            Transaction_Amount_Deviation,
            Days_Since_Last_Transaction,
            amount,
            output
        ]
    )

demo.launch(share=True)
```
